```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

t = np.linspace(-10,10,1000)
narr = np.arange(20)+1

temp = t*0.0
```

```
for n in narr:
  temp = temp +(1-(-1)**n)/(n*np.pi)*np.sin(2*np.pi*n*t)
  

f, (ax1) = plt.subplots(1, figsize=(12,9))

ax1.plot(t, temp)
  
```
